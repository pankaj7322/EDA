## importing the data of country code and country

```python
df_country = pd.read_excel('Country-Code.xlsx')
df_country.head()
```

```python
df.columns
```

```python
## meging the two dataset into one
final_df = pd.merge(df,df_country, on = 'Country Code', how = 'left')
```

```python
final_df.head()
```

```python
final_df.head(2)
```

```python
final_df.dtypes
```

```python
final_df.columns
```

```python
country_names = final_df.Country.value_counts().index
country_val = final_df.Country.value_counts().values
```

```python
## Pie chart top 3  countries that uses zomato
plt.pie(country_val[:3], labels= country_names[:3], autopct='%1.2f%%' )
```

#### observation
- Zomato maximum records or transaction from india after that USA and Then United Kingdoms

```python
final_df.columns
```

```python
ratings = final_df.groupby(['Aggregate rating','Rating color','Rating text']).size().reset_index().rename(columns = {0:'Rating Count'})
```

```python
ratings
```

## Observation
1. When Rating is between 4.5 to 4.9 -----> Excellent
2. When Rating is between 4.0 to 4.4 -----> Very Good
3. When Rating is between 3.5 to 3.9 -----> Good
4. When Rating is between 3.0 to 3.4 -----> Average
5. When Rating is between 2.0 to 2.4 -----> Poor

```python
ratings.head()
```

```python
sns.barplot(x = 'Aggregate rating', y ='Rating Count', data = ratings)
```

```python
sns.barplot(x="Aggregate rating", y="Rating Count", hue = "Rating color", data = ratings, palette = ['blue','red','orange','green','darkgreen'])
```

## Observation
1. Not rated count is very high
2. Maximum number of rating are between 2.5 to 3.4

```python
## count plot
sns.countplot(x = 'Rating color', data = ratings , palette = ['blue','red','orange','green','springgreen'])
```
